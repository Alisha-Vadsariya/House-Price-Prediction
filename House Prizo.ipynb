```python
def convertRange(x):
    
    temp = x.split('_')
    if len(temp) == 2:
        return(float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        
        st = np.std(subdf.price_per_sqft)
        
        gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index = True)
    return df_output
df_train = remove_outliers_sqft(df_train)
df_train.describe()
```

```python
def BHK_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        BHK_stats = {}
        for BHK, BHK_df in location_df.groupby('BHK'):
            BHK_stats[BHK] = {
                'mean': np.mean(BHK_df.price_per_sqft),
                'std' : np.std(BHK_df.price_per_sqft),
                'count': BHK_df.shape[0]
            }
        for BHK, BHK_df in location_df.groupby('BHK'):
            stats = BHK_stats.get(BHK-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, BHK_df[BHK_df.price_per_sqft<(stats['mean'])].index.vlaues)
            return df.drop(exclude_indices,axis='index')
```

```python
df_train = BHK_outlier_remover(df_train)
```

# Linear Regression

```python
#Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#Create a column transformer that applies one-hot encoding to the 'location' column and passes through the remaining columns
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['location']),remainder = 'passthrough')
```

```python
#Create a scaler to standardize the input features
scaler = StandardScaler()
```

```python
#Create a linear regression model with normalization
lr = LinearRegression(normalize=True)
```

```python
#Create a pipeline that applies the column transformer, scaler, and linear regression model in sequence
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
#Fit the pipeline to the training data
pipe.fit(X_train,y_train)
```

```python
#Make predictions on the test data using the fitted pipeline
y_pred_lr = pipe.predict(X_test)
```

```python
#Compute the R-squared score on the test data
r2_score(y_test,y_pred_lr)
```

# Lasso

```python
#import lasso regression model
lasso = Lasso()
```

```python
#create pipeline to preprocess data and apply lasso regression
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
#Fit pipeline on training data
pipe.fit(X_train,y_train)
```

```python
#Use pipeline to make predictions on test data
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso) #calculate R-squared score
```

# Ridge 

```python
#Define ridge regression model
ridge = Ridge()
```

```python
#create a pipeline that applies the column transformation step, scaling step, and then the ridge regression model
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
#fit the pipeline on the training data
pipe.fit(X_train,y_train)
```

```python
#make predictions on the test data using the pipeline
y_pred_ridge = pipe.predict(X_test)
#calculate the R-squared score on the test data using the ridge regression model
r2_score(y_test,y_pred_ridge)
```

```python
#calculate the R-squared score on the test data using the simple linear regression model
print("No Regularization: ",r2_score(y_test,y_pred_lr))
#Calculate the R-squared score on the test data using the Lasso regression model
print("Lasso: ",r2_score(y_test,y_pred_lasso))
#calculate the R-squared score on the test data using the lasso regression model
print("Ridge: ",r2_score(y_test,y_pred_ridge))
```
